#                                                      PYTHON DATA TYPE CATEGORIES 


## Basic Types:

Int: Whole nums (1, 2, 3).  
Float: Decimals (3.14, 2.5).  
Complex: Real+Imag (1 + 2j).  
Bool: True/False.  
Str: Text ("hello").  

## Container Types:
List: Ordered items.  
[1, 2, 3]
Tuple: Ordered, immutable.  
(1, 2, 3)
Set: Unordered, unique.  
{1, 2, 3}
Dict: Key-value pairs.  
{'k1': 'v1', 'k2': 'v2'}

## User-Defined Types:
Class: Obj blueprint.

# int

```python
# Numbers
print(4)

# Large Numbers
print(1e308)  # 1 * 10^308
print(1e309)  # 'inf' (exceeds float limit) BUT BELOW THIS RANGE IT CAN EASILY HANDLE INTEGERS 
# PYTHON CAN HANDLE BIG RANGE OF INTEGERS ALSO.
```

# FLOAT

```python
# Floating-Point Numbers
print(4.5)  # Standard

# Large Floating-Point Numbers
print(1.8e307)  # Large
print(1.8e308)  # Larger
print(1.9e308)  # Too large (inf) Usually, we see that these much larger numbers of data don't exist in real time, but for an exceptional case, we can go ahead
```

# BOOL 

```python
print("True")
print("False")
```

# complex

```python
# IN PYTHON ONLY J WORKS AS IN A COMPLEX DATA TYPE.
print(4+9j) #Python uses j (engineering notation) instead of i (math notation) for imaginary numbers..
```

# STRING

```python
# Strings with Different Quotation Styles
print('Piyush Rawat')       # single quotes
print("Piyush Rawat")       # double quotes
print('''Piyush Rawat''')   # triple single quotes
print("""Piyush Rawat""")   # triple double quotes
```
